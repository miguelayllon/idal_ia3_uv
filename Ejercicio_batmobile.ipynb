```
class Fuel:
    def __init__(self):
      self.numFuel = 1200 #litros

    #Metodos generales
    def stock_Fuel(self):
      y = self.numFuel
      return y

    def subtract_Fuel(self,delta):
      self.numFuel = self.numFuel - delta
      y = self.numFuel
      return y

    def add_Fuel(self,delta):
      self.numFuel = self.numFuel + delta 
      y = self.numFuel
      return y      
```

```
class Pem:
    def __init__(self):
      self.numPem = 6

    #Metodos generales
    def stock_Pem(self):
      y = self.numPem
      return y

    def subtract_Pem(self,delta):
      self.numPem = self.numPem - delta
      y = self.numPem
      return y

    def add_Pem(self,delta):
      self.numPem = self.numPem + delta 
      y = self.numPem
      return y
```

```
class Rocket:
    def __init__(self):
      self.numRocket = 8

    #Metodos generales
    def stock_Rocket(self):
      y = self.numRocket
      return y

    def subtract_Rocket(self,delta):
      self.numRocket = self.numRocket - delta
      y = self.numRocket
      return y

    def add_Rocket(self,delta):
      self.numRocket = self.numRocket + delta 
      y = self.numRocket
      return y
```
